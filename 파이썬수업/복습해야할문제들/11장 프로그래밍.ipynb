```python
# 1번
from tkinter import *

root = Tk()

def ch():
    l1['text'] = "clicked"



l1 = Label(root,text="Hi!!")
l1.grid(row=0,column=0)
b1 = Button(root,text="click me",command=ch)
b1.grid(row=0,column=1)

root.mainloop()
```

```python
# 4번

from tkinter import *

root = Tk()

text1 = ["이름","주소","도","우편번호","국가"]

for i in range(5):
        l1 = Label(root,text=f"{text1[i]}:")
        l1.grid(row=i,column=0)
        e1 = Entry(root,width=50)
        e1.grid(row=i,column=1)


b1 = Button(root,text="clear",width=10)
b1.grid(row=6,column=1 ,sticky="e",padx=5,pady=10)
b2 = Button(root,text="submit",width=10)
b2.grid(row=6,column=2,sticky="e",padx=10,pady=10)
root.mainloop()
```

```python
# 5번

from tkinter import*
root = Tk()

def up() :
    total = int(l1["text"])
    total += 1
    l1["text"] = str(total)
    
def down() :
    total = int(l1["text"])
    if total > 0:
        total -= 1
        l1["text"] = str(total)


b1 = Button(root,text="감소",command=down)
b1.grid(row=0,column=0)
l1 = Label(root,text="0")
l1.grid(row=0,column=1)
b2 = Button(root,text="증가",command=up)
b2.grid(row=0,column=3)
root.mainloop()
```

```python
# 8번
from tkinter import*

root = Tk()

def log() :
    print("로그인 완료")

def reset() :
    e1.delete(0,END)
    e2.delete(0,END)




l1 = Label(root,text="아이디")
l1.grid(row=0,column=0)

l2 = Label(root,text="패스워드")
l2.grid(row=1,column=0)

e1 = Entry(root)
e1.grid(row=0,column=1)

e2 = Entry(root)
e2.grid(row=1,column=1)

b1 = Button(root,text="로그인",command=log)
b1.grid(row=2,column=0)

b2 = Button(root,text="취소", command=reset)
b2.grid(row=2,column=1, sticky="w")



root.mainloop()
```
