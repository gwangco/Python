```python
# 2번

class Box :
    def __init__(self,l,h,d):
        self.__l = l
        self.__h = h 
        self.__d = d
    
    def setHeight(self,h):
        self.__h = h
        
    def getHeight(self) :
        return self.__h
    
    def setLength(self,l) :
        self.__l = l
        
    def getLength(self) :
        return self.__l
    
    def setDepth(self,d) :
        self.__d = d
        
    def getDepth(self) :
        return self.__d
    
    def __str__(self):
        return f"({self.__l},{self.__h},{self.__d})"

b1 = Box(100,100,100)
print(b1)
print("상자의 부피는", b1.getHeight()*b1.getLength()*b1.getDepth())
```

```python
# 3번

class Dog :
    count = 0
    def __init__(self,name,age,color):
        self.name = name
        self.age = age
        self.color = color
        Dog.count += 1
        self.number = Dog.count
        
    def __str__(self):
        return f'강아지 #{self.number} = "{self.name}",{self.age},"{self.color}"'

co1 = Dog("Molly",10,"brown")
co2 = Dog("Daisy",6,"black")
co3 = Dog("Bella",7,"white")

print(co1)
print(co2)
print(co3)
print("현재까지 생성된 강아지의 수 = ", + Dog.count)

```

```python
# 4번


class Dog :
    count = 0
    def __init__(self,name,age,color):
        self.name = name
        self.age = age
        self.color = color
        Dog.count += 1
        self.number = Dog.count
        
    def __str__(self):
        return f'강아지 #{self.number} = "{self.name}",{self.age},"{self.color}"'
    
    def younger(self, other_dog):
        other_dog.age -= 5  # 나이를 5살 감소
        
co1 = Dog("Molly",10,"brown")
co2 = Dog("Daisy",6,"black")
co3 = Dog("Bella",7,"white")

co1.younger(co1)
co2.younger(co2)
co3.younger(co3)

print(co1)
print(co2)
print(co3)
print("현재까지 생성된 강아지의 수 = ", + Dog.count)
```

```python
# 5번

class Circle :
    count = 0
    def __init__(self, radius):
        self.radius = radius
        Circle.count += 1
        self.number = Circle.count

    # __gt__ 메서드: 반지름 비교
    def __gt__(self, other):
        return self.radius > other.radius

    def __str__(self):
        return f"원 #{self.number} = 반지름 {self.radius}"

# Circle 객체 생성
circle1 = Circle(10)
circle2 = Circle(20)
print(circle1)
print(circle2)

# 비교 및 출력
if circle1 > circle2:
    print("원 #1 > 원 #2는 True 입니다.")
else:
    print("원 #1 > 원 #2는 False 입니다.")
```

```python
# 6번 


class Person:
    count = 0
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.count += 1
        self.number = Person.count

    def show(self):
        print(f"이름: {self.name}, 나이: {self.age}")


class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    
    def show(self):
        print(f"직장인 #{self.number} = 이름: {self.name}, 나이: {self.age}, 연봉: {self.salary}만원")



emp = Employee("Kim", 27, 8000)
emp.show()

    
```

```python
# 7번

class Account:
    def __init__(self, user_id, password):
        self.user_id = user_id
        self.password = password

    def __str__(self):
        return f"아이디={self.user_id} 패스워드={self.password}"



accounts = []

while True:
    print("\n1. 아이디 등록하기")
    print("2. 로그인 하기")
    print("3. 모든 사용자 아이디 출력")
    print("4. 종료")
    choice = int(input("메뉴를 입력하세요: "))

    if choice == 1: 
        user_id = input("아이디 입력: ")
        password = input("패스워드 입력: ")
        accounts.append(Account(user_id, password))
        print("등록 완료!")

    elif choice == 2:  
        login_id = input("아이디 입력: ")
        login_pw = input("패스워드 입력: ")
       
        found = False
        for account in accounts:
            if account.user_id == login_id and account.password == login_pw:
                print("로그인 성공!")
                found = True
                break
        if not found:
            print("로그인 실패: 아이디 또는 패스워드가 잘못되었습니다.")

    elif choice == 3:  
        print("현재 등록된 사용자:")
        for i, account in enumerate(accounts, start=1):
            print(f"사용자 #{i} {account}")

    elif choice == 4:  
        print("프로그램을 종료합니다.")
        break

    else:
        print("잘못된 입력입니다. 다시 선택해주세요.")
```
