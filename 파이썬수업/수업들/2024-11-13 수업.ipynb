```python
from tkinter import *

root = Tk()
root.geometry("600x400")

def aaa():
    if man.get() == 2:
        c4.pack_forget()
        c5.pack_forget()
    if man.get() == 1:
        c4.pack(side=LEFT)
        c5.pack(side=LEFT)

def menu():
    temp = ""
    if m1.get() == 1:
        temp += "아메리카노(2000) "
    if m2.get() == 1:
        temp += "라떼(3000) "
    if m3.get() == 1:
        temp += "콜라(500) "
    if m4.get() == 1:
        temp += "맥주(4000) "
    if m5.get() == 1:
        temp += "소주(3000) "
    l2["text"] = temp

def price():
    total = 0
    if m1.get() == 1:
        total += 2000
    if m2.get() == 1:
        total += 3000
    if m3.get() == 1:
        total += 500
    if m4.get() == 1:
        total += 4000
    if m5.get() == 1:
        total += 3000
    e3.delete(0, END)
    e3.insert(0, f"총액: {total}원")

def pay():
    p1.pack()

def clear(payment_type):
    e4.pack()
    if payment_type == "현금":
        e4.delete(0, END)
        e4.insert(0, "현금 결제가 완료됐습니다.")
    elif payment_type == "카드":
        e4.delete(0, END)
        e4.insert(0, "카드 결제가 완료됐습니다.")

l1 = Label(root, text="우리카페")
l1.pack()

Label(root, text=" ").pack()

f2 = Frame(root)
man = IntVar()

r1 = Radiobutton(f2, text="성인", variable=man, value=1, command=aaa)
r1.pack(side=LEFT)
r2 = Radiobutton(f2, text="청소년", variable=man, value=2, command=aaa)
r2.pack(side=LEFT)

f2.pack()
Label(root, text=" ").pack()

f1 = Frame(root)

m1 = IntVar()
m2 = IntVar()
m3 = IntVar()
m4 = IntVar()
m5 = IntVar()

c1 = Checkbutton(f1, text="아메리카노(2000원)", variable=m1, command=menu)
c1.pack(side=LEFT)
c2 = Checkbutton(f1, text="라떼(3000원)", variable=m2, command=menu)
c2.pack(side=LEFT)
c3 = Checkbutton(f1, text="콜라(500원)", variable=m3, command=menu)
c3.pack(side=LEFT)
c4 = Checkbutton(f1, text="맥주(4000원)", variable=m4, command=menu)
c4.pack(side=LEFT)
c5 = Checkbutton(f1, text="소주(3000원)", variable=m5, command=menu)
c5.pack(side=LEFT)

f1.pack()
Label(root, text=" ").pack()

l2 = Label(root, text="메뉴를 선택하세요")
l2.pack()

b1 = Button(root, text="가격 계산", command=price)
b1.pack()

e3 = Entry(root)
e3.pack()

b2 = Button(root, text="결제", command=pay)
b2.pack()

p1 = Frame(root)
b3 = Button(p1, text="현금", command=lambda: clear("현금"))
b4 = Button(p1, text="카드", command=lambda: clear("카드"))
b3.pack(side=LEFT)
b4.pack(side=LEFT)

e4 = Entry(root)

root.mainloop()

```
