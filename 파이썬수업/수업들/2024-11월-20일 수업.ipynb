```python
# 퀴즈 교수님 답                          #자동화로 메뉴 만드는거 만들어보기

from tkinter import *
root = Tk()
root.geometry("600x400")


def order_menu() :
    s = 0
    msg = ""
    if e1.get() != "" :
        msg  = msg + "아메리카노" + e1.get() + "잔. "
        s = s + int(e1.get()) * 3000
    if e2.get() != "" :
        msg  = msg + "라떼" + e2.get() + "잔. "
        s = s + int(e2.get()) * 2000
    if e3.get() != "" :
        msg  = msg + "콜라" + e3.get() + "잔. "
        s = s + int(e3.get()) * 1000
    l1["text"] = msg + "=>" + str(s) + "원. "
    
    

Label(root, text="까페").pack()

f1 = Frame(root)
Label(f1, text="아메리카노(3000원)").pack(side=LEFT)
e1 = Entry(f1)
e1.pack(side=LEFT)
Label(f1,text="잔").pack(side=LEFT)
f1.pack()

f2 = Frame(root)
Label(f2, text="라떼(2000원)").pack(side=LEFT)
e2 = Entry(f2)
e2.pack(side=LEFT)
Label(f2,text="잔").pack(side=LEFT)
f2.pack()

f3 = Frame(root)
Label(f3, text="콜라(1000원)").pack(side=LEFT)
e3 = Entry(f3)
e3.pack(side=LEFT)
Label(f3,text="잔").pack(side=LEFT)
f3.pack()

b1 = Button(root,text="주문하기", command=order_menu)
b1.pack()

l1 = Label(root,text="주문하세요")
l1.pack()

root.mainloop()
```

```python
# 퀴즈 자동화

from tkinter import*

root = Tk()
root.geometry("600x400")
Label(text="까페테리아 메뉴").pack()

def menu_order() :
    total = 0
    order = ""
    for i,j in enumerate(ml):
        num = entries[i].get()
        if num != "" :
            order = order + f"{j} {num}잔, "
            total = total + int(num) * p1[i]
        l1["text"] = f"{order} => {total}원"
                

ml = ["아메리카노","라떼","카페모카","바닐라라떼","자몽주스"]
p1 = [3000,2000,1000,4000,5000]

entries = []

for i,j in enumerate(ml):
    frame = Frame(root)
    Label(frame,text=f"{j}{p1[i]}원").pack(side=LEFT)
    entry = Entry(frame)
    entry.pack(side=LEFT)
    Label(frame,text="잔").pack(side=LEFT)
    frame.pack()
    entries.append(entry)

b1 = Button(root,text="주문하기",command=menu_order)
b1.pack()
l1 = Label(root,text="주문하세요")
l1.pack()

root.mainloop()
```
