```python
# dropdown menu 만들기                             #혼자서 완성본 만들어보기
from tkinter import *
from tkinter import filedialog
from PIL import ImageTk

root = Tk()
root.geometry("600x600")

img =NONE

c = Canvas(root, width=600, height=600)
c.pack()
    

def fopen() :
    global img
    fn = filedialog.askopenfilename(parent=root, filetypes=(("jpg 파일", "*.jpg"),("png 파일", "*.png"),("모든 파일", "*.*")))
    img = ImageTk.PhotoImage(file=fn)
    c.create_image(30,30, anchor=NW, image = img)

def new_file():
    global img, current_image
    c.delete("all")
    img = None
    current_image = None

def save_file():
    if current_image:
        save_path = filedialog.asksaveasfilename(
            defaultextension=".png", 
            filetypes=(("PNG 파일", "*.png"), ("JPEG 파일", "*.jpg"), ("모든 파일", "*.*")),
            title="저장 위치 선택 및 파일 이름 입력"
        )
        if save_path:
            current_image.save(save_path)
    

m1 = Menu(root)

m2 = Menu(m1, tearoff=0)
m2.add_command(label="새파일", command= new_file)
m2.add_command(label="파일열기", command=fopen)
m2.add_command(label="저장", command=save_file)
m2.add_separator()
m2.add_command(label="끝내기", command=root.destroy)   # 주피터 노트북에서는 quit이 사용이 안됌 그러나 단일창에서는 사용가능(주피터에서destroy는 사용가능)

m1.add_cascade(label="파일", menu=m2)

root.config(menu=m1)

root.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog
from PIL import Image, ImageTk

root = Tk()
root.geometry("600x600")

img = None
current_image = None

c = Canvas(root, width=600, height=600)
c.pack()

def fopen():
    global img, current_image
    fn = filedialog.askopenfilename(
        parent=root, 
        filetypes=(("jpg 파일", "*.jpg"), ("png 파일", "*.png"), ("모든 파일", "*.*"))
    )
    if fn:
        current_image = Image.open(fn)
        img = ImageTk.PhotoImage(current_image)
        c.delete("all")
        c.create_image(30, 30, anchor=NW, image=img)

def new_file():
    global img, current_image
    c.delete("all")
    img = None
    current_image = None

def save_file():
    if current_image:
        save_path = filedialog.asksaveasfilename(
            defaultextension=".png", 
            filetypes=(("PNG 파일", "*.png"), ("JPEG 파일", "*.jpg"), ("모든 파일", "*.*")),
            title="저장 위치 선택 및 파일 이름 입력"
        )
        if save_path:
            current_image.save(save_path)

# 메뉴 설정
m1 = Menu(root)

m2 = Menu(m1, tearoff=0)
m2.add_command(label="새파일", command=new_file)  # 새 파일
m2.add_command(label="파일 열기", command=fopen)  # 파일 열기
m2.add_command(label="저장", command=save_file)   # 파일 저장
m2.add_separator()
m2.add_command(label="끝내기", command=root.destroy)  # 프로그램 종료

m1.add_cascade(label="파일", menu=m2)

root.config(menu=m1)
root.mainloop()
```
